```python
from ultralytics import YOLO
import cv2
import numpy as np

# Load the YOLO model
model = YOLO("best.pt")

# Predict on the image
results = model.predict(source="test1.jpeg", conf=0.6)

# Load the image
image = cv2.imread("test1.jpeg")

# Extract boxes and classes
boxes = results.xyxy[0][:, :4].cpu().numpy().astype(int)
classes = results.xyxy[0][:, 5].cpu().numpy().astype(int)

# Create an empty mask
mask = np.zeros_like(image)

# Draw masks for each object
for box, cls in zip(boxes, classes):
    x1, y1, x2, y2 = box
    class_color = (0, 255, 0)  # Green color for masks, you can change this if you like
    cv2.rectangle(mask, (x1, y1), (x2, y2), class_color, -1)

# Apply the mask to the image
masked_image = cv2.bitwise_and(image, mask)

# Show the original image and the masked image
cv2.imshow("Original Image", image)
cv2.imshow("Masked Image", masked_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
